# Fly data download

Please follow the instructions on the README to download the data from [Zenodo](https://zenodo.org/records/11002776) and store it in the './data/fly/' directory. 


```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

with h5py.File('../data/fly/Fly_DLC_behavior_tracking.h5', 'r') as f:
    data = f['data'][:]

# shape samples x sequence length x keypoints x 2
assert data.shape == (28059, 234, 64), "data shape is not correct"
print("Download successful!")

# frame rate
fps = 80 # Hz 
time = np.arange(0, data.shape[1]/fps, 1/fps)
```

### Visualise the time series for individual keypoints

```python
fig, ax = plt.subplots(2, 1, figsize=(5,4), sharex=True)
ax[0].plot(time, data[0, :, 32])
ax[0].set_ylabel('keypoint 32')
ax[1].plot(time, data[0, :, 18])
ax[1].set_xlabel('time (s)')
ax[1].set_ylabel('keypoint 18')
plt.tight_layout()

fig, ax = plt.subplots(2, 1, figsize=(5,4), sharex=True)
ax[0].plot(time, data[0, :, 33])
ax[0].set_ylabel('keypoint 33')
ax[1].plot(time, data[0, :, 19])
ax[1].set_ylabel('keypoint 19')
ax[1].set_xlabel('time (s)')
plt.tight_layout()
```

### Plot one time point x,y coordinates (fly head pointing downwards)

```python
sample_idx = 0
time_point = 0
plt.figure(figsize=(3,3))
# even indices are x coordinates, odd indices are y coordinates
plt.plot(data[sample_idx, time_point, 0::2], data[sample_idx, time_point, 1::2], 'o', ms=4)
plt.xlabel('x')
plt.ylabel('y')
```
